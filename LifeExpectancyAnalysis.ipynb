```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("nailasrivastava/life-expectancy-analysis")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing the Kaggle data set - Life-Expectancy-Analysis

```python
df = pd.read_csv(path + "/Life Expectancy Data.csv")
df.head()
print(df.columns)
```

Dropped any rows with NA

```python
df = df.dropna()
df.head()
```

Dropped 'Thinness 1-19 years' and 'Thinness 5-9 years'

```python
df = df.drop(columns=[' thinness  1-19 years', ' thinness 5-9 years'], axis=1)
df.head()
```

```python
df['Life expectancy '] = df['Life expectancy '].astype(int)
df['Adult Mortality'] = df['Adult Mortality'].astype(int)
df['Polio'] = df['Polio'].astype(int)
df['Diphtheria '] = df['Diphtheria '].astype(int)
df['Population'] = df['Population'].astype(int)
df['Polio'] = df['Polio'].astype(int)
df['Hepatitis B'] = df['Hepatitis B'].astype(int)
df = df.rename(columns={"infant deaths": "Infant Deaths", 'under-five deaths ': 'Under 5 Deaths',
                        'Life expectancy ': 'Life Expectancy', 'Income composition of resources': 'ICR',
                        "percentage expenditure": "Expenditure%"})
df.GDP = df.GDP.round(3)
df['Expenditure%'] = df['Expenditure%'].round(3)

df.head()
```

```python
num_of_countries = df['Country'].nunique()
print("Number of countries:", num_of_countries)
```
